```python
import math
def homo_average(img, mask, point, T):
    av_val = img[mask > 0].sum() / np.count_nonzero(img[mask > 0])
                                                            
    if abs(av_val - img[point]) <= T:
        return True
    
    return False
    
```

```python
def region_growing(image, seed_point,homo_fun,r, T):
    mask = np.zeros(image_gray.shape, np.uint8)
    mask[seed_point] = 1
    count = 1
    while count > 0:
        count = 0
        local_mask = np.zeros(image_gray.shape, np.uint8)
        for i in range(r,image.shape[0] - r):
            for j in range(r,image.shape[1] - r):
                if mask[i,j]==0 and mask[i - r:i + r, j-r: j+r].sum() > 0:
                    if homo_fun(image, mask, (i,j), T):
                        local_mask[i,j] = 1
        count = np.count_nonzero(local_mask)
        print(count)
        mask += local_mask
        
    return mask*255
    
```

```python
seed_point = (250,250)
mask = region_growing(image_gray,seed_point,homo_average,1, 10)
```

```python
plt.imshow(mask, cmap="gray")
```

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('sar_1.jpg')
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
import math
def homo_average(img, mask, point, T):
    av_val = img[mask > 0].sum() / np.count_nonzero(img[mask > 0])

    if abs(av_val - img[point]) <= T:
        return True

    return False
```

```python
def region_growing(image, seed_point,homo_fun,r, T):
    mask = np.zeros(image_gray.shape, np.uint8)
    mask[seed_point] = 1
    count = 1
    while count > 0:
        count = 0
        local_mask = np.zeros(image_gray.shape, np.uint8)
        for i in range(r,image.shape[0] - r):
            for j in range(r,image.shape[1] - r):
                if mask[i,j]==0 and mask[i - r:i + r, j-r: j+r].sum() > 0:
                    if homo_fun(image, mask, (i,j), T):
                        local_mask[i,j] = 1
        count = np.count_nonzero(local_mask)
        print(count)
        mask += local_mask

    return mask*255
```

```python
flags = cv2.KMEANS_RANDOM_CENTERS
image_gray_copy = image_gray.copy()
z = image_gray_copy.reshape((-1,3))
z = np.float32(z)
criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 10, 1.0)
K = 3
ret,label,center=cv2.kmeans(z,K,None,criteria,10,cv2.KMEANS_RANDOM_CENTERS)
center = np.uint8(center)
res = center[label.flatten()]
res2 = res.reshape((image_gray_copy.shape))
plt.imshow(res2, cmap="gray")
```

```python
image = cv2.imread('oranges_1.jpg')
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

plt.imshow(image_gray, cmap="gray")
```

```python
plt.figure(figsize = (15,10))

ret, thresh = cv2.threshold(image_gray,0,255, cv2.THRESH_BINARY_INV+cv2.THRESH_OTSU)
plt.subplot(2, 4, 1)
plt.imshow(thresh, cmap="gray")
plt.axis('off')

dist = cv2.distanceTransform(thresh, cv2.DIST_L2, 5)
plt.subplot(2, 4, 2)
plt.imshow(dist, cmap="gray")
plt.axis('off')

ret, sure_fg = cv2.threshold(dist, 0.1 * dist.max(), 255, cv2.THRESH_BINARY)
plt.subplot(2, 4, 3)
plt.imshow(sure_fg, cmap="gray")
plt.axis('off')

sure_fg = sure_fg.astype(np.uint8)
ret, markers = cv2.connectedComponents(sure_fg)
plt.subplot(2, 4, 4)
plt.imshow(markers, cmap="gray")
plt.axis('off')
```

```python
image = cv2.imread('sar_1.jpg')
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

seed_point = (250, 250)
mask_average = region_growing(image_gray, seed_point, homo_average, 4, 25)
plt.imshow(mask_average, cmap="gray")
plt.axis('off')
```

```python
def homo_median(img, mask, point, T):
    region = np.median(img[mask > 0])
    return abs(img[point] - region) <= T

seed_point = (250, 250)
mask_median = region_growing(image_gray, seed_point, homo_median, 4, 25)
plt.imshow(mask_median, cmap="gray")
plt.axis('off')
```

```python
image = cv2.imread('palm_1.JPG')
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

blurred = cv2.GaussianBlur(image_gray, (13, 13), 0)

ret, thresh = cv2.threshold(blurred,0,255, cv2.THRESH_BINARY+cv2.THRESH_OTSU)

kernel = np.ones((3,3),np.uint8)
thresh = cv2.dilate(thresh, kernel, iterations=2)
thresh = cv2.erode(thresh, kernel, iterations=2)
```

```python
plt.figure(figsize = (15,10))

plt.subplot(2, 4, 1)
plt.imshow(thresh, cmap="gray")
plt.axis('off')

dist = cv2.distanceTransform(thresh, cv2.DIST_L2, 5)
plt.subplot(2, 4, 2)
plt.imshow(dist, cmap="gray")
plt.axis('off')

ret, sure_fg = cv2.threshold(dist, 0.12 * dist.max(), 255, cv2.THRESH_BINARY)
plt.subplot(2, 4, 3)
plt.imshow(sure_fg, cmap="gray")
plt.axis('off')

sure_fg = np.uint8(sure_fg)
ret, markers = cv2.connectedComponents(sure_fg)
plt.subplot(2, 4, 4)
plt.imshow(markers)
plt.axis('off')

print(f"- : {ret - 2}")
```
